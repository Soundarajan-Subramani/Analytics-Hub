```python
# Finding the upper and lower bound values

def finding_range(x):
    xs = pd.Series(sorted(x))
    Q3 = xs.quantile(q=0.75)
    Q1 = xs.quantile(q=0.25)
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
outliers = col

for i in outliers:
    lr,ur = finding_range(df[i])
    print("Variable: {}".format(i))
    print("")
    print("Upper Bound Value: {}".format(round(ur,2)))
    print("Lower Bound Value: {}".format(round(lr,2)))
    print("\n")
```

```python
# treating slag variable
df["slag"] = np.where(df["slag"]>356.25, 356.25, df["slag"])

# treating water variable variable
df["water"] = np.where(df["water"]>232.35, 232.35, df["water"])
df["water"] = np.where(df["water"]<127.15, 127.15, df["water"])

# treating super_plast variable
df["super_plast"] = np.where(df["super_plast"]>25, 25, df["super_plast"])

# treating fine_agg variable
df["fine_agg"] = np.where(df["fine_agg"]>969.05, 969.05, df["fine_agg"])

# treating age variable
df["age"] = np.where(df["age"]>129.5, 129.5, df["age"])
```

### Modelling Prerequisites

```python
# dependent and independent variables

x = df.drop("strength", axis=1)
y = df["strength"]
```

```python
# train and test splits
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# checking the train and test record counts
print("No.of.records in train data: {}".format(x_train.shape[0]))
print("No.of.records in test data: {}".format(x_test.shape[0]))
```

```python
# scaled version of training and testing data

x_train_sc = x_train.apply(zscore)
x_test_sc = x_test.apply(zscore)
```

```python
# decision tree regressor-------------------------------------(raw model)

model = DecisionTreeRegressor(random_state=1).fit(x_train,y_train)

# accuracy-----------------------------------------------------

train_score = model.score(x_train,y_train)      # or we can use r2_score(y_train, train_predictions)
test_score  = model.score(x_test,y_test)        # or we can use r2_score(y_test, test_predictions)

# rmse and mape values 

train_predictions = model.predict(x_train)
test_predictions  = model.predict(x_test)

train_rmse = np.sqrt(mse(y_train,train_predictions))
test_rmse  = np.sqrt(mse(y_test,test_predictions))

train_mae = mae(y_train,train_predictions)
test_mae  = mae(y_test,test_predictions)

print("Training Data------------------------")
print("")
print("Model Score: {}".format(round(train_score,2)))
print("RMSE Value : {}".format(round(train_rmse,2)))
print("MAPE Value : {}".format(round(train_mae,2)))
print("\n")
print("Testing Data-------------------------")
print("")
print("Model Score: {}".format(round(test_score,2)))
print("RMSE Value : {}".format(round(test_rmse,2)))
print("MAPE Value : {}".format(round(test_mae,2)))
```

```python
# random forest regressor-------------------------------------(raw model)

model = RandomForestRegressor(random_state=1).fit(x_train,y_train)

# evaluation metrics-----------------------------------------------------

train_score = model.score(x_train,y_train)
test_score  = model.score(x_test,y_test)

train_predictions = model.predict(x_train)
test_predictions  = model.predict(x_test)

train_rmse = np.sqrt(mse(y_train,train_predictions))
test_rmse  = np.sqrt(mse(y_test,test_predictions))

train_mae = mae(y_train,train_predictions)
test_mae  = mae(y_test,test_predictions)

print("Training Data------------------------")
print("")
print("Model Score: {}".format(round(train_score,2)))
print("RMSE Value : {}".format(round(train_rmse,2)))
print("MAPE Value : {}".format(round(train_mae,2)))
print("\n")
print("Testing Data-------------------------")
print("")
print("Model Score: {}".format(round(test_score,2)))
print("RMSE Value : {}".format(round(test_rmse,2)))
print("MAPE Value : {}".format(round(test_mae,2)))
```

```python
# linear regressor--------------------------------------------(raw model)

model = LinearRegression().fit(x_train,y_train)

# evaluation metrics-----------------------------------------------------

train_score = model.score(x_train,y_train)
test_score  = model.score(x_test,y_test)

train_predictions = model.predict(x_train)
test_predictions  = model.predict(x_test)

train_rmse = np.sqrt(mse(y_train,train_predictions))
test_rmse  = np.sqrt(mse(y_test,test_predictions))

train_mae = mae(y_train,train_predictions)
test_mae  = mae(y_test,test_predictions)

print("Training Data------------------------")
print("")
print("Model Score: {}".format(round(train_score,2)))
print("RMSE Value : {}".format(round(train_rmse,2)))
print("MAPE Value : {}".format(round(train_mae,2)))
print("\n")
print("Testing Data-------------------------")
print("")
print("Model Score: {}".format(round(test_score,2)))
print("RMSE Value : {}".format(round(test_rmse,2)))
print("MAPE Value : {}".format(round(test_mae,2)))
```

```python
# ann regressor-----------------------------------------------(raw model)

model = MLPRegressor(hidden_layer_sizes=(500), random_state=1).fit(x_train_sc,y_train)

# evaluation metrics-----------------------------------------------------

train_score = model.score(x_train_sc,y_train)
test_score  = model.score(x_test_sc,y_test)

train_predictions = model.predict(x_train_sc)
test_predictions  = model.predict(x_test_sc)

train_rmse = np.sqrt(mse(y_train,train_predictions))
test_rmse  = np.sqrt(mse(y_test,test_predictions))

train_mae = mae(y_train,train_predictions)
test_mae  = mae(y_test,test_predictions)

print("Training Data------------------------")
print("")
print("Model Score: {}".format(round(train_score,2)))
print("RMSE Value : {}".format(round(train_rmse,2)))
print("MAPE Value : {}".format(round(train_mae,2)))
print("\n")
print("Testing Data-------------------------")
print("")
print("Model Score: {}".format(round(test_score,2)))
print("RMSE Value : {}".format(round(test_rmse,2)))
print("MAPE Value : {}".format(round(test_mae,2)))
```

### Tuned Decision Tree Regressor

```python
# using gridsearch to find the best parameters

parameters = {'max_depth':[8,10,12],
              'min_samples_leaf':[1,2,3],
              'min_samples_split':[5,10,15]}

model = DecisionTreeRegressor(random_state=1)
grid  = GridSearchCV(estimator=model, param_grid=parameters, cv=5).fit(x_train, y_train)

print(grid.best_params_)
```

### Performance Comparison

```python
dtr_model = DecisionTreeRegressor(max_depth=10, min_samples_leaf=2,min_samples_split=5, random_state=1).fit(x_train,y_train)
ann_model = MLPRegressor(activation='tanh', hidden_layer_sizes=(100, 100), solver='adam').fit(x_train_sc,y_train)
rfr_model = RandomForestRegressor(max_features=5, n_estimators=100, random_state=1).fit(x_train,y_train)
lnr_model = LinearRegression().fit(x_train,y_train)
```

```python
# predictions and evaluation (tuned decision tree regressor)

# model scores-------------------------------------------
dtr_train_score = round(dtr_model.score(x_train,y_train),4)      
dtr_test_score  = round(dtr_model.score(x_test,y_test),4)        

# tuned decision tree regressor--------------------------
dtr_train_predictions = dtr_model.predict(x_train)
dtr_test_predictions = dtr_model.predict(x_test)

# root mean squared error--------------------------------
dtr_train_rmse = np.sqrt(mse(y_train,dtr_train_predictions))
dtr_test_rmse  = np.sqrt(mse(y_test,dtr_test_predictions))

# mean absolute error------------------------------------
dtr_train_mae = mae(y_train,dtr_train_predictions)
dtr_test_mae  = mae(y_test,dtr_test_predictions)
```
